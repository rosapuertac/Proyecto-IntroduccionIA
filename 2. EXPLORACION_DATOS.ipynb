# **Analisis estadístico**

```python
Tabletm=pd.DataFrame()
dic={ "Media":float(dataset[['total_deaths']].mean()), "Mediana":float(dataset[['total_deaths']].median()), "Desviación_estándar":float(dataset[['total_deaths']].std())}
Tabletm=Tabletm.append(dic,ignore_index=True)
Tabletm.style.hide_index()
```

```python
Tablets=pd.DataFrame()
dic={ "Media":float(dataset[['total_cases']].mean()), "Mediana":float(dataset[['total_cases']].median()), "Desviación_estándar":float(dataset[['total_cases']].std())}
Tablets=Tablets.append(dic,ignore_index=True)
Tablets.style.hide_index()
```

```python
Tablenw=pd.DataFrame()
dic={ "Media":float(dataset[['new_cases']].mean()), "Mediana":float(dataset[['new_cases']].median()), "Desviación_estándar":float(dataset[['new_cases']].std())}
Tablenw=Tablenw.append(dic,ignore_index=True)
Tablenw.style.hide_index()
```

```python
#Genero los diagramas de barras
#Para total de muertos
histogram2= Tabletm.plot.hist()
print(histogram2)
plt.show()

#Para total de casos
histogram1= Tablets.plot.hist()
print(histogram1)
plt.show()


#Para los nuevos casos
histogram = Tablenw.plot.hist()
print(histogram)
plt.show()
```

```python
#Número de registros por pais
paises=[]
cantidad=[]

#Recorro los paises
for pais in dataset["location"].unique().tolist():
  paises.append(pais)
  cantidad.append(dataset[dataset["location"]==pais].count()[0])

#Ahora genero la tabla de pandas
countries = pd.DataFrame(data={'Pais': paises, 'Cantidad de datos': cantidad})
countries
```

```python
Tablec=pd.DataFrame()
dic={ "Media":float(countries[['Cantidad de datos']].mean()), "Mediana":float(countries[['Cantidad de datos']].median()), "Desviación_estándar":float(countries[['Cantidad de datos']].std())}
Tablec=Tablec.append(dic,ignore_index=True)
Tablec.style.hide_index()
```

```python
#Selecciono el que más datos tenga
countries[countries["Cantidad de datos"]==np.max(countries["Cantidad de datos"])]
```
